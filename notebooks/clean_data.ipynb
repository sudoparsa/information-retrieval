```python
import gc
import pandas as pd
from tqdm import tqdm
import numpy as np
import json
tqdm.pandas()
```

```python
df = pd.read_json('resources/data.json', orient='index')
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
def dropna_references(references):
    res = []
    for dic in references:
        if dic['paperId'] != None and dic['authors'] != None:
            res.append(dic)
    return res
def dropna_authors(authors):
    res = []
    for dic in authors:
        if dic['authorId'] != None and dic['name'] != None:
            res.append(dic)
    return res
df['references'] = df['references'].apply(dropna_references)
df['authors'] = df['authors'].apply(dropna_authors)
df.head()
```

```python
a = set()
for id, fields in df['fieldsOfStudy'].items():
    for field in fields:
        a.add(field)
for field in a:
    print(field, sum(df['fieldsOfStudy'].apply(lambda x: field in x)))
```

```python
df = df[df['fieldsOfStudy'].apply(str).str.contains('Computer Science') | 
df['fieldsOfStudy'].apply(str).str.contains('Mathematics')]
df.shape
```

```python
df['citationCount'].describe()
```

```python
df = df[df['citationCount'] > 50]
df.shape
```

```python
def get_k_first_references(references, k=10):
    if len(references) <= k:
        return references
    return references[:k]
df['k_references'] = df['references'].apply(get_k_first_references)
df.head()
```

```python
df.drop(columns=['fieldsOfStudy', 'references'], inplace=True)
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```
